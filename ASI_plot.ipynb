```python
import pandas as pd
import matplotlib.pyplot as plt

# Read and clean the data
df = pd.read_csv("data.csv")
df2 = df.dropna(subset=['p21003_i0', 'p21021_i0'])

plt.figure(figsize=(8, 6))
plt.scatter(df2['p21003_i0'], df2['p21021_i0'], label='Data', alpha=0.6)

# Compute the median ASI for each age
median_by_age = df2.groupby('p21003_i0')['p21021_i0'].median().reset_index()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(median_by_age['p21003_i0'], median_by_age['p21021_i0'],
         marker='o', color='red', label='Median ASI')
plt.xlabel('Age')
plt.ylabel('Arterial Stiffness Index')
plt.title('Median ASI by Age')
plt.legend()
plt.show()
```
